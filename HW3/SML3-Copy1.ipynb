```python
def gradient_descent(X,y,theta,alpha,n):
    
    m = len(y)
        
    for i in range(n):
        prediction = np.dot(X,theta)
        h = 1 / (1 + np.exp(-prediction))
        theta = theta - (1/m) * alpha * (X.T.dot(h - y))
        
    return theta
```
