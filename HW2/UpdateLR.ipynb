```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

from sklearn import linear_model
from sklearn.metrics import mean_squared_error
from sklearn import preprocessing
```

```python
# read in data
us_youtube = pd.read_csv('us_youtube_videos.csv')
us_youtube['publish_time'] = pd.to_datetime(us_youtube['publish_time'], format="%Y-%m-%d %H:%M:%S")
```

```python
us_youtube
```

```python
# creating publishing time dummy variable
us_youtube['published_in_morning'] = us_youtube['publish_time'].apply(lambda x: x.hour >= 12)
us_youtube['published_in_morning'] = np.where(us_youtube['published_in_morning'], 1, 0)
```

```python
# get only features to use
us_youtube = us_youtube[['views', 'likes', 'dislikes', 'published_in_morning', 'comment_count']]
us_youtube = us_youtube.reset_index(drop=True)
```

```python
# min max normalize features
min_max_scaler = preprocessing.MinMaxScaler()
us_youtube['views'] = min_max_scaler.fit_transform(pd.DataFrame(us_youtube['views']))
us_youtube['likes'] = min_max_scaler.fit_transform(pd.DataFrame(us_youtube['likes']))
us_youtube['dislikes'] = min_max_scaler.fit_transform(pd.DataFrame(us_youtube['dislikes']))
us_youtube['comment_count'] = min_max_scaler.fit_transform(pd.DataFrame(us_youtube['comment_count']))
```

```python
us_youtube.head(15)
```

```python
us_youtube.describe()
```

```python
views_comments = us_youtube[['views', 'comment_count']]
```

```python
reg = linear_model.LinearRegression()
reg.fit(views_comments.drop(columns='comment_count'), views_comments['comment_count'])

comments_pred = reg.predict(views_comments.drop(columns='comment_count'))

plt.scatter(views_comments['views'], views_comments['comment_count'], color='black')
plt.plot(views_comments['views'], comments_pred, color='blue', linewidth=3)
plt.xlabel('Views')
plt.ylabel('Number of Comments')
plt.show()
```

```python
print("views coef: {}".format(reg.coef_[0]))
```

```python
comments_pred
```

```python
mse_single = mean_squared_error(views_comments['comment_count'], comments_pred)
print("MSE: {}".format(mse_single))
print("RSE: {}".format(mse_single ** 0.5))
```

```python
reg_multi = linear_model.LinearRegression()
reg_multi.fit(us_youtube.drop(columns='comment_count'), us_youtube['comment_count'])
```

```python
comments_pred_multi = reg_multi.predict(us_youtube.drop(columns='comment_count'))
```

```python
feature_names = us_youtube.drop(columns='comment_count').columns
for i in range(len(reg_multi.coef_)):
    print("{} coef: {}".format(feature_names[i], reg_multi.coef_[i]))
```

```python
mse_multi = mean_squared_error(us_youtube['comment_count'], comments_pred_multi)
print("MSE: {}".format(mse_multi))
print("RSE: {}".format(mse_multi ** 0.5))
```
