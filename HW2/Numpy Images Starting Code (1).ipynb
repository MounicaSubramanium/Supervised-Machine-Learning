### Matrix Manipulation Visualization

```python
import numpy as np
from PIL import Image
import requests
from io import BytesIO
import matplotlib.pyplot as plt

response = requests.get("https://i.imgur.com/LEMM22X.jpg")
img = np.array(Image.open(BytesIO(response.content)))

%matplotlib inline

def show(np_img):
    plt.imshow(np_img)
    plt.show()
```

```python
print(img.shape)  

show(img)

```
